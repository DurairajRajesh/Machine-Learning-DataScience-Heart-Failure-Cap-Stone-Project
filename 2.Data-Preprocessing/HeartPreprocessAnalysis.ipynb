```python
def quanqual(df):
    quan=[]
    qual=[]
    for columnName in df.columns:
        if df[columnName].dtype == 'O':
            qual.append(columnName)
        else:
            quan.append(columnName)
    return quan, qual
```

```python
quan,qual=quanqual(df)
```

```python
df[quan]
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(df[quan])
datf = imp.transform(df[quan])
datf = pd.DataFrame(datf,columns = quan)
```

```python
datf
```

```python
categ=df[qual]
```

```python
categ
```

```python
two=[datf,categ]
```

```python
preprocessed=pd.concat(two,axis=1)
```
